# **LEARN STATISTICS WITH PYTHON: Life Expectancy by Country**

## Step 1: Import Libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Step 2: Load and View the dataset

```python
data = pd.read_csv("country_data.csv")
```

```python
data.head()
```

## Step 3: Isolate the life expectancy column

```python
life_expectancy = data["Life Expectancy"]
```

## Step 4: Find quartiles of life expectancy

```python
life_expectancy_quartiles = np.quantile(life_expectancy, [0.25, 0.5, 0.75])
print("Life Expectancy Quartiles:", life_expectancy_quartiles)
```

## Step 5: Any Plot of life expectancy

Histogram with better binning and visualization

```python
plt.figure(figsize=(8, 5))
plt.hist(life_expectancy, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Life Expectancy")
plt.xlabel("Life Expectancy (years)")
plt.ylabel("Frequency")
plt.grid("True", linestyle='--', alpha=0.6)
```

Vertical lines for quartiles

```python
q1, q2, q3 = life_expectancy_quartiles
plt.axvline(q1, color='red', linestyle='dashed', linewidth=1, label='Q1')
plt.axvline(q2, color='green', linestyle='dashed', linewidth=1, label='Q2 (Median)')
plt.axvline(q3, color='blue', linestyle='dashed', linewidth=1, label='Q3')
plt.legend()
plt.tight_layout()
plt.show()
```

Boxplot (Visualizes Quartiles Clearly)

```python
plt.figure(figsize=(6, 4))
plt.boxplot(life_expectancy, vert=False)
plt.title("Boxplot of Life Expectancy")
plt.xlabel("Life Expectancy (years)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

Violin Plot (More detailed than boxplot)

```python
import seaborn as sns

plt.figure(figsize=(6, 4))
sns.violinplot(x=life_expectancy, color="lightblue")
plt.title("Violin Plot of Life Expectancy")
plt.xlabel("Life Expectancy (years)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

Histogram with All Quartile Lines + Mean


```python
mean_val = life_expectancy.mean()

plt.figure(figsize=(8, 5))
plt.hist(life_expectancy, bins=10, color='lightgray', edgecolor='black')
plt.axvline(q1, color='red', linestyle='dashed', linewidth=1.5, label='Q1')
plt.axvline(q2, color='green', linestyle='dashed', linewidth=1.5, label='Q2 (Median)')
plt.axvline(q3, color='blue', linestyle='dashed', linewidth=1.5, label='Q3')
plt.axvline(mean_val, color='orange', linestyle='solid', linewidth=1.5, label='Mean')

plt.title("Life Expectancy with Quartiles and Mean")
plt.xlabel("Life Expectancy (years)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

## Step 6: Determine quartile for a life expectancy value

```python
life_expectancy_value = 70

# Use previously computed quartiles
q1, q2, q3 = life_expectancy_quartiles

if life_expectancy_value < q1:
    print(f"A life expectancy of {life_expectancy_value} years falls in the 1st quarter.")
elif life_expectancy_value < q2:
    print(f"A life expectancy of {life_expectancy_value} years falls in the 2nd quarter.")
elif life_expectancy_value < q3:
    print(f"A life expectancy of {life_expectancy_value} years falls in the 3rd quarter.")
else:
    print(f"A life expectancy of {life_expectancy_value} years falls in the 4th quarter.")    
```

## Step 7: Isolate GDP column

```python
gdp = data["GDP"]
```

## Step 8: Find the median GDP

```python
median_gdp = np.median(gdp)
print("Median GDP:", median_gdp)
```

## Step 9: Split data based on GDP

```python
low_gdp = data[data["GDP"] <= median_gdp]
low_gdp
```

```python
high_gdp = data[data["GDP"] > median_gdp]
high_gdp
```

## Step 10: Quartiles for low GDP countries

```python
low_gdp_quartiles = np.quantile(low_gdp["Life Expectancy"], [0.25, 0.5, 0.75])
print("Low GDP Life Expectancy Quartiles:", low_gdp_quartiles)
```

## Step 11: Quartiles for high GDP countries

```python
high_gdp_quartiles = np.quantile(high_gdp["Life Expectancy"], [0.25, 0.5, 0.75])
print("Low GDP Life Expectancy Quartiles:", high_gdp_quartiles)
```

## Step 12: Plot histograms for comparison

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Histograms for both groups
plt.hist(high_gdp["Life Expectancy"], bins=10, alpha=0.5, label='High GDP', color='lightgreen', edgecolor='black')
plt.hist(low_gdp["Life Expectancy"], bins=10, alpha=0.5, label='Low GDP', color='salmon', edgecolor='black')

# Labels and styling
plt.title("Life Expectancy: High vs Low GDP Countries")
plt.xlabel("Life Expectancy (years)")
plt.ylabel("Frequency")
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### High vs Low GDP + Quartile Lines

```python
# Unpack quartiles
lq1, lq2, lq3 = low_gdp_quartiles
hq1, hq2, hq3 = high_gdp_quartiles

plt.figure(figsize=(10, 6))

# Histograms
plt.hist(high_gdp["Life Expectancy"], bins=10, alpha=0.5, label='High GDP', color='lightgreen', edgecolor='black')
plt.hist(low_gdp["Life Expectancy"], bins=10, alpha=0.5, label='Low GDP', color='salmon', edgecolor='black')

# Vertical Lines for High GDP Quartiles
plt.axvline(hq1, color='darkgreen', linestyle='dashed', linewidth=1.5, label='High GDP Q1')
plt.axvline(hq2, color='green', linestyle='dashed', linewidth=1.5, label='High GDP Median')
plt.axvline(hq3, color='lightgreen', linestyle='dashed', linewidth=1.5, label='High GDP Q3')

# Vertical Lines for Low GDP Quartiles
plt.axvline(lq1, color='darkred', linestyle='dashed', linewidth=1.5, label='Low GDP Q1')
plt.axvline(lq2, color='red', linestyle='dashed', linewidth=1.5, label='Low GDP Median')
plt.axvline(lq3, color='salmon', linestyle='dashed', linewidth=1.5, label='Low GDP Q3')

# Labels, grid, and legend
plt.title("Life Expectancy by GDP Group with Quartiles")
plt.xlabel("Life Expectancy (years)")
plt.ylabel("Frequency")
plt.legend(loc='upper left', fontsize=9)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

## Step 13: Interpretation Interpretation for 70 Years Life Expectancy

```python
# Interpretation
life_expectancy_value = 70

# Unpack Quartiles
lq1, lq2, lq3 = low_gdp_quartiles
hq1, hq2, hq3 = high_gdp_quartiles

# Function to determine quartile

def find_quartile(value, q1, q2, q3):
    if value < q1:
        return "1st Quartile"
    elif value < q2:
        return "2nd Quartile"
    elif value < q3:
        return "3rd Quartile"
    else:
        return "4th Quartile"
    
# Interpret for Low GDP
low_gdp_position = find_quartile(life_expectancy_value, lq1, lq2, lq3)
print(f"In LOW GDP countries, a life expectancy of {life_expectancy_value} years falls in the {low_gdp_position}.")

# Interpret for High GDP
high_gdp_position = find_quartile(life_expectancy_value, hq1, hq2, hq3)
print(f"In HIGH GDP countries, a life expectancy of {life_expectancy_value} years falls in the {high_gdp_position}.")
```

## Final Conclusion:

This analysis shows a clear correlation between a country's GDP and its average life expectancy. Countries with higher GDPs tend to have significantly higher life expectancy quartiles, suggesting better healthcare, living conditions, and public services.
